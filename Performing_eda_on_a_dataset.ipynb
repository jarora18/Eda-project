```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
df = pd.read_csv("iris.data", header=None, names=column_names)

# Basic Info
print("Dataset Info:")
print(df.info())
print("\nFirst 5 Rows:")
print(df.head())
print("\nStatistical Summary:")
print(df.describe())
print("\nClass Distribution:")
print(df['class'].value_counts())

# Pairplot
sns.pairplot(df, hue="class", markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Features", y=1.02)
plt.show()

# Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Matrix of Features")
plt.show()

# Boxplots
plt.figure(figsize=(12, 8))
for i, column in enumerate(column_names[:-1]):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x="class", y=column, data=df)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()

# Histograms
df.hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("Histograms of Each Feature", y=1.02)
plt.tight_layout()
plt.show()
```
